```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
log = pd.read_csv("../build/log.csv")
log.head()
```

```python
fig = plt.figure(figsize=(9, 6))
plt.xlabel("time step, k")
plt.ylabel("Cross track error")
plt.grid()
plt.plot(log["cte"], '-', color="c",label="cte")
plt.hlines(0,0,len(log["cte"]), color="r",label="control")
plt.legend(loc=1) 
plt.title("Cross Track Error")
plt.show()
```

```python
fig = plt.figure(figsize=(9, 6))
plt.xlabel("time step, k")
plt.ylabel("Speed (mph)")
plt.grid()
plt.plot(log["speed"], '-', color="c",label="speed")
plt.plot(log["desired_speed"], '-', color="r",label="control")
plt.legend(loc=1) 
plt.title("Speed")
plt.show()
```
